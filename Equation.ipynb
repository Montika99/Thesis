```python
I = symbols('I', cls=Function)
a, b, t, v, theta, m =symbols('a b t v theta m')
I_2 = [Eq(I(t).diff(t), -a-b*t-theta*I(t))]
#ics = {I(v): 0}
dsolve(I_2)
```

```python
C2 = ((a*exp(v*theta))/theta-(b*t*exp(v*theta))/theta-(b*exp(v*theta))/theta**2)/(exp(-v*theta))
display(C2)
```

```python
#I_2t = C2*exp(-t*theta)-(a*exp(t*theta))/theta+(b*t*exp(t*theta))/theta-(b*exp(t*theta))/theta**2
#display(I_2t)
#simI_2 = simplify(I_2t)
#display(simI_2)
#print(simI_2)
I_2t = (exp(-t*theta)*(a*theta*(exp(theta*v)-exp(theta*t))+b*exp(theta*t)*(1-theta*t)+b*exp(theta*v)*(theta*v-1)))/theta**2
display(I_2t)
```

```python
I_2t = (exp(-m*theta)*(a*theta*(exp(theta*v)-exp(theta*m))+b*exp(theta*m)*(1-theta*m)+b*exp(theta*v)*(theta*v-1)))/theta**2
#I2_m =(-b*exp(2*m*theta) + theta*(-a + b*m)*exp(2*m*theta) + (-b + theta*(a - b*m))*exp(2*theta*v))*exp(-m*theta)/theta**2
display(I2_m)
```

```python
I = symbols('I', cls=Function)
a, b, t, v, theta, m =symbols('a b t v theta m')
I_1 = [Eq(I(t).diff(t), -a-b*t)]
#ics = {I(v): 0}
dsolve(I_1)
```

```python
C1 = I2_m+a*t+(b*t)/2
display(C1)
```

```python
I_1t = C1 - a*t - (b*t**2)/2
print(I_1t)
display(I_1t)
```

Total Cost

-Purchasing cost

```python
qi = (-b*exp(2*m*theta) + theta*(-a + b*m)*exp(2*m*theta) + (-b + theta*(a - b*m))*exp(2*theta*v))*exp(-m*theta)/theta**2
display(qi)
demand = a+b*t
intqs = integrate(demand, (t, v, T))
print(intqs)
qs = T**2*b/2 + T*a - a*v - b*v**2/2
Q = qs+qi
```

```python
PC = p*(qi+qs)
```

-Holding cost

```python
intI1 = integrate(I_1t, (t,0,m))
display(intI1)
intI2 = integrate(I_2t, (t,m,v))
display(intI2)
print(intI2)
```

```python
simintI2 = simplify(-((-a*theta**6 + b*m*theta**6 - 2*b*theta**5)*exp(m*theta) + (-a*theta**6*exp(2*theta*v) + b*m*theta**6*exp(2*theta*v) + 2*b*theta**5*exp(2*theta*v))*exp(-m*theta))/theta**8 + ((-a*theta**6 + b*theta**6*v - 2*b*theta**5)*exp(theta*v) + (-a*theta**6*exp(2*theta*v) + b*theta**6*v*exp(2*theta*v) + 2*b*theta**5*exp(2*theta*v))*exp(-theta*v))/theta**8)
print(simintI2)
display(simintI2)
```

```python
HC = h*(intI1+simintI2)
print(HC)
display(HC)
simhc = simplify(HC)
display(simhc)
```

-Shortage cost

```python
SC = s*qs
display(SC)
```

-Deterioration cost 

```python
demandof0tov = integrate(demand, (t,0,v))
print(demandof0tov)
display(demandof0tov)
noofdeter = Q-(a*v + b*v**2/2)
DC = noofdeter*p
display(DC)
```

-Odering cost

```python
RC = r
```

-Total cost

```python
TC = 1/T*(PC + HC + SC + DC + RC)
print(TC)
display(TC)
```

-DiffTotalCost
